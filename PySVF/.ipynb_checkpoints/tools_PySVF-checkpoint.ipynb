````python
def Index1D(i,j,k,N):
  return int((k*N + j)*N +i)
````

````python
def PeriodicGrid( i : int, N : int):
  ip = i
  if (i >= N): 
     ip = i - N	  
  if (i < 0):
     ip = i + N	  
  return ip
````
